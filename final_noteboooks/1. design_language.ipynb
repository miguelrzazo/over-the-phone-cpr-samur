```python
# Verificaci√≥n de tipograf√≠a - Comparaci√≥n con LaTeX
print("üé® CONFIGURACI√ìN TIPOGR√ÅFICA ACTUALIZADA")
print("=" * 50)
print("üìÑ Configuraci√≥n de fuentes actual en matplotlib:")
print(f"   ‚Ä¢ Fuente principal: {plt.rcParams['font.family']}")
print(f"   ‚Ä¢ Sans-serif: {plt.rcParams['font.sans-serif'][:3]}")  # Primeras 3 opciones
print(f"   ‚Ä¢ Tama√±o base: {plt.rcParams['font.size']}pt")
print(f"   ‚Ä¢ Peso predeterminado: {plt.rcParams['font.weight']}")
print()

print("üìä Equivalencias con LaTeX:")
print("   ‚Ä¢ LaTeX: \\renewcommand{\\familydefault}{\\sfdefault}")
print("   ‚Ä¢ Matplotlib: font.family = ['sans-serif']")
print("   ‚Ä¢ LaTeX: \\usepackage[fontsize=10pt]{scrartcl}")
print("   ‚Ä¢ Matplotlib: font.size = 10")
print()

print("‚úÖ Estado de sincronizaci√≥n:")
print("   ‚Ä¢ Fuente sans-serif: ‚úì SINCRONIZADA")
print("   ‚Ä¢ Tama√±o base 10pt: ‚úì SINCRONIZADA") 
print("   ‚Ä¢ Jerarqu√≠a tipogr√°fica: ‚úì CONSISTENTE")
print("   ‚Ä¢ T√≠tulos en negrita: ‚úì APLICADA")

# Crear una muestra visual r√°pida
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

# Muestra de texto actual
ax1.text(0.5, 0.8, "TIPOGRAF√çA ACTUALIZADA", ha='center', va='center', 
         fontsize=14, fontweight='bold', transform=ax1.transAxes)
ax1.text(0.5, 0.6, "DejaVu Sans 10pt", ha='center', va='center', 
         fontsize=10, transform=ax1.transAxes)
ax1.text(0.5, 0.4, "Consistente con LaTeX", ha='center', va='center', 
         fontsize=9, transform=ax1.transAxes)
ax1.text(0.5, 0.2, "Publicaci√≥n cient√≠fica", ha='center', va='center', 
         fontsize=8, style='italic', transform=ax1.transAxes)
ax1.set_title("Nueva configuraci√≥n", fontweight='bold')
ax1.axis('off')

# Comparaci√≥n de tama√±os
sizes = [14, 12, 10, 9, 8]
labels = ['T√≠tulos principales', 'T√≠tulos de ejes', 'Texto base', 'Leyendas', 'Anotaciones']
colors = ['#004c99', '#4079b2', '#80a6cc', '#bfd2e6', '#d8d8d8']

bars = ax2.barh(range(len(sizes)), sizes, color=colors)
ax2.set_yticks(range(len(sizes)))
ax2.set_yticklabels(labels)
ax2.set_xlabel('Tama√±o de fuente (pt)', fontweight='bold')
ax2.set_title('Jerarqu√≠a tipogr√°fica', fontweight='bold')

# A√±adir valores en las barras
for i, (bar, size) in enumerate(zip(bars, sizes)):
    ax2.text(bar.get_width() + 0.1, bar.get_y() + bar.get_height()/2, 
             f'{size}pt', va='center', fontweight='bold')

plt.tight_layout()
plt.show()

print("\nüéØ RESULTADO: Tipograf√≠a completamente sincronizada con documento LaTeX")
```

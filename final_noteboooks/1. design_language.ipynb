# Nuevo Lenguaje de Dise√±o Visual - RCP Transtelef√≥nica
## Paleta de Azules y Coral

Este notebook establece el nuevo lenguaje de dise√±o visual para el estudio, utilizando una paleta de azules con un toque de coral para destacar. El objetivo es crear visualizaciones modernas, claras y consistentes.

### Objetivos del Dise√±o
- **Consistencia tipogr√°fica** con documentos LaTeX.
- **Nueva paleta crom√°tica** basada en azules y coral.
- **Jerarqu√≠a visual** clara y profesional.
- **Legibilidad** optimizada para publicaci√≥n.

# ‚ö†Ô∏è Aviso Importante

Este notebook utiliza **datos simulados (mock data)** √∫nicamente con fines de demostraci√≥n del dise√±o gr√°fico. Los datos no representan resultados reales del estudio.

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd

# Configuraci√≥n optimizada
plt.rcParams.update({
    'font.family': 'sans-serif',
    'font.size': 10,
    'axes.titlesize': 12,
    'figure.dpi': 300,
    'savefig.dpi': 300,
    'figure.facecolor': 'white',
})

# Paleta crom√°tica consistente (azules y gris, naranja solo para destacar)
COLORS = {
    'azul_principal': '#304C89',      # Azul principal
    'azul_oscuro': '#2E282A',        # Azul muy oscuro (casi negro, para ejes/texto)
    'azul_medio': '#58A4B0',         # Azul medio
    'gris_claro': '#E4E6C3',         # Gris claro de fondo
    'naranja': '#FA8334',            # Naranja solo para destacar
    # Tonos alternativos derivados de la paleta base
    'azul_claro': '#7FC7D9',         # Alternativa clara a 58A4B0
    'azul_pastel': '#B7D6E8',        # Azul pastel para fondos suaves
    'gris_medio': '#B0B3A1',         # Gris medio para l√≠neas o detalles
    'negro': '#181A1B',              # Para texto o contraste fuerte
    'blanco': '#FFFFFF',             # Fondo puro
}

# Datos base para optimizar velocidad
np.random.seed(42)
BASE_DATA = {
    'edades': np.random.normal(65, 15, 100),
    'tiempos': np.random.normal(10, 3, 100),
    'outcomes': np.random.binomial(1, 0.4, 100),
    'meses': np.arange(1, 13),
}

print("üé® Lenguaje de Dise√±o - 25 Tipos de Gr√°ficos")
print("üìä Paleta: Azules/gris, naranja solo para destacar")
print("‚ö° Configuraci√≥n optimizada")
```

```python
# Grid Principal (5x5) - Parte 1: Gr√°ficos 1-12
fig1 = plt.figure(figsize=(20, 15))
gs1 = fig1.add_gridspec(3, 4, hspace=0.4, wspace=0.3)
fig1.suptitle('Galer√≠a de Visualizaciones - Parte 1 (1-12)', fontsize=16, fontweight='bold')

# 1. Barras con IC
ax = fig1.add_subplot(gs1[0, 0])
ax.bar(['Telef√≥nica', 'Presencial'], [0.42, 0.38], yerr=[0.08, 0.07], 
       color=[COLORS['azul_principal'], COLORS['azul_claro']], capsize=4)
ax.set_title('1. Barras con IC', fontweight='bold')
ax.set_ylabel('Tasa ROSC')
ax.grid(True, alpha=0.3)

# 2. Box Plot
ax = fig1.add_subplot(gs1[0, 1])
bp = ax.boxplot([BASE_DATA['tiempos'][:50], BASE_DATA['tiempos'][50:]], patch_artist=True)
bp['boxes'][0].set_facecolor(COLORS['azul_principal'])
bp['boxes'][1].set_facecolor(COLORS['azul_claro'])
ax.set_xticklabels(['Telef√≥nica', 'Presencial'])
ax.set_title('2. Box Plot', fontweight='bold')
ax.grid(True, alpha=0.3)

# 3. Histograma
ax = fig1.add_subplot(gs1[0, 2])
ax.hist(BASE_DATA['edades'], bins=15, color=COLORS['azul_medio'], alpha=0.7)
ax.axvline(np.mean(BASE_DATA['edades']), color=COLORS['naranja'], linestyle='--', lw=2)
ax.set_title('3. Histograma', fontweight='bold')
ax.grid(True, alpha=0.3)

# 4. Dispersi√≥n
ax = fig1.add_subplot(gs1[0, 3])
x = BASE_DATA['tiempos'][:50]
y = 0.6 - 0.02 * x + np.random.normal(0, 0.1, 50)
ax.scatter(x, y, color=COLORS['azul_claro'], alpha=0.6)
z = np.polyfit(x, y, 1)
ax.plot(x, np.poly1d(z)(x), color=COLORS['naranja'], linestyle='--')
ax.set_title('4. Dispersi√≥n', fontweight='bold')
ax.grid(True, alpha=0.3)

# 5. Barras Apiladas
ax = fig1.add_subplot(gs1[1, 0])
grupos = ['<65', '65-75', '>75']
rosc = [45, 42, 35]
no_rosc = [55, 58, 65]
ax.bar(grupos, rosc, color=COLORS['azul_oscuro'], label='ROSC')
ax.bar(grupos, no_rosc, bottom=rosc, color=COLORS['azul_claro'], label='No ROSC')
ax.set_title('5. Barras Apiladas', fontweight='bold')
ax.legend()
ax.grid(True, alpha=0.3)

# 6. L√≠neas Temporales
ax = fig1.add_subplot(gs1[1, 1])
tel = 0.4 + 0.01 * BASE_DATA['meses'] + np.random.normal(0, 0.02, 12)
pres = 0.35 + 0.005 * BASE_DATA['meses'] + np.random.normal(0, 0.02, 12)
ax.plot(BASE_DATA['meses'], tel, 'o-', color=COLORS['azul_principal'], label='Telef√≥nica')
ax.plot(BASE_DATA['meses'], pres, 's-', color=COLORS['naranja'], label='Presencial')
ax.set_title('6. Series Temporales', fontweight='bold')
ax.legend()
ax.grid(True, alpha=0.3)

# 7. Forest Plot (CORREGIDO)
ax = fig1.add_subplot(gs1[1, 2])
vars = ['Edad >65', 'Sexo M', 'Comorbilidad']
ors = [1.2, 0.8, 1.5]
ci_l = [0.9, 0.6, 1.1]
ci_u = [1.6, 1.1, 2.0]
y_pos = range(len(vars))
for i, (or_val, lower, upper) in enumerate(zip(ors, ci_l, ci_u)):
    color = COLORS['azul_principal'] if or_val > 1 else COLORS['naranja']
    ax.errorbar(or_val, i, xerr=[[or_val-lower], [upper-or_val]], 
               fmt='o', color=color, ecolor=color, capsize=4, markersize=6)
ax.axvline(1, color=COLORS['gris_medio'], linestyle='--')
ax.set_yticks(y_pos)
ax.set_yticklabels(vars)
ax.set_title('7. Forest Plot', fontweight='bold')
ax.grid(True, alpha=0.3)

# 8. Curva ROC
ax = fig1.add_subplot(gs1[1, 3])
fpr = np.linspace(0, 1, 50)
tpr = 1 - (1 - fpr)**2
ax.plot(fpr, tpr, color=COLORS['azul_principal'], lw=2, label='AUC = 0.75')
ax.plot([0, 1], [0, 1], color=COLORS['gris_medio'], linestyle='--')
ax.fill_between(fpr, tpr, alpha=0.2, color=COLORS['azul_pastel'])
ax.set_title('8. Curva ROC', fontweight='bold')
ax.legend()
ax.grid(True, alpha=0.3)

# 9. Violin Plot
ax = fig1.add_subplot(gs1[2, 0])
data_v = [BASE_DATA['edades'][:33], BASE_DATA['edades'][33:66], BASE_DATA['edades'][66:]]
parts = ax.violinplot(data_v, positions=[1, 2, 3])
for i, pc in enumerate(parts['bodies']):
    pc.set_facecolor([COLORS['azul_principal'], COLORS['azul_claro'], COLORS['azul_pastel']][i])
    pc.set_alpha(0.7)
ax.set_xticks([1, 2, 3])
ax.set_xticklabels(['ROSC', 'No ROSC', '√âxitus'])
ax.set_title('9. Violin Plot', fontweight='bold')
ax.grid(True, alpha=0.3)

# 10. Heatmap
ax = fig1.add_subplot(gs1[2, 1])
corr_data = np.random.rand(4, 4)
corr_data = (corr_data + corr_data.T) / 2
np.fill_diagonal(corr_data, 1)
from matplotlib.colors import ListedColormap
azul_cmap = ListedColormap([
    COLORS['azul_oscuro'], COLORS['azul_medio'], COLORS['azul_principal'], COLORS['azul_claro'], COLORS['azul_pastel']
    ])
im = ax.imshow(corr_data, cmap=azul_cmap, vmin=0, vmax=1)
ax.set_xticks(range(4))
ax.set_yticks(range(4))
ax.set_xticklabels(['Edad', 'Tiempo', 'ROSC', 'CPC'])
ax.set_yticklabels(['Edad', 'Tiempo', 'ROSC', 'CPC'])
ax.set_title('10. Heatmap', fontweight='bold')

# 11. √Årea
ax = fig1.add_subplot(gs1[2, 2])
x = BASE_DATA['meses']
y1 = 0.4 + 0.01 * x + np.random.normal(0, 0.02, 12)
y2 = 0.35 + 0.005 * x + np.random.normal(0, 0.02, 12)
ax.fill_between(x, y1, alpha=0.5, color=COLORS['azul_principal'], label='Telef√≥nica')
ax.fill_between(x, y2, alpha=0.5, color=COLORS['azul_claro'], label='Presencial')
ax.set_title('11. Gr√°fico de √Årea', fontweight='bold')
ax.legend()
ax.grid(True, alpha=0.3)

# 12. Dot Plot
ax = fig1.add_subplot(gs1[2, 3])
hospitales = ['H1', 'H2', 'H3', 'H4']
tasas = [0.42, 0.38, 0.45, 0.35]
media = np.mean(tasas)
colors_d = [COLORS['azul_principal'], COLORS['azul_claro'], COLORS['azul_pastel'], COLORS['gris_medio']]
ax.scatter(tasas, hospitales, c=colors_d, s=80)
ax.axvline(media, color=COLORS['gris_medio'], linestyle='--')
ax.set_title('12. Dot Plot', fontweight='bold')
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# Grid Principal - Parte 2: Gr√°ficos 13-25
fig2 = plt.figure(figsize=(20, 16))
gs2 = fig2.add_gridspec(4, 4, hspace=0.4, wspace=0.3)
fig2.suptitle('Galer√≠a de Visualizaciones - Parte 2 (13-25)', fontsize=16, fontstyle='italic')

# 13. Barras Horizontales
ax = fig2.add_subplot(gs2[0, 0])
factores = ['Edad', 'Tiempo\nresp.', 'Comorbilidad']
import_vals = [0.8, 0.65, 0.45]
ax.barh(factores, import_vals, color=COLORS['azul_principal'])
ax.set_title('13. Barras Horizontales', fontstyle='italic')
ax.grid(True, alpha=0.3)

# 14. Radar Chart
ax = fig2.add_subplot(gs2[0, 1], projection='polar')
cats = ['ROSC', 'Supervivencia', 'CPC Fav.', 'Tiempo']
vals_t = [0.8, 0.6, 0.5, 0.7]
vals_p = [0.7, 0.7, 0.6, 0.5]
angles = np.linspace(0, 2*np.pi, len(cats), endpoint=False).tolist()
vals_t += vals_t[:1]
vals_p += vals_p[:1]
angles += angles[:1]
ax.plot(angles, vals_t, color=COLORS['naranja'], lw=2, label='Telef√≥nica')
ax.plot(angles, vals_p, color=COLORS['azul_principal'], lw=2, label='Presencial')
ax.fill(angles, vals_t, alpha=0.25, color=COLORS['naranja'])
ax.fill(angles, vals_p, alpha=0.25, color=COLORS['azul_claro'])
ax.set_xticks(angles[:-1])
ax.set_xticklabels(cats)
ax.set_title('14. Radar Chart', fontstyle='italic')

# 15. Pie Chart
ax = fig2.add_subplot(gs2[0, 2])
sizes = [35, 65]
colors_pie = [COLORS['azul_principal'], COLORS['azul_claro']]
wedges, texts, autotexts = ax.pie(sizes, labels=['ROSC', 'No ROSC'], 
                                 autopct='%1.1f%%', colors=colors_pie)
ax.set_title('15. Gr√°fico Circular', fontstyle='italic')

# 16. Stem Plot
ax = fig2.add_subplot(gs2[0, 3])
x_stem = range(5)
y_stem = [0.35, 0.45, 0.65, 0.8, 0.42]
markerline, stemlines, baseline = ax.stem(x_stem, y_stem)
markerline.set_markerfacecolor(COLORS['naranja'])
markerline.set_markeredgecolor(COLORS['naranja'])
stemlines.set_color(COLORS['azul_principal'])
ax.set_title('16. Stem Plot', fontstyle='italic')
ax.grid(True, alpha=0.3)

# 17. Step Plot
ax = fig2.add_subplot(gs2[1, 0])
x_step = BASE_DATA['meses']
y_step = np.cumsum(np.random.normal(0.4, 0.1, 12))
ax.step(x_step, y_step, where='mid', color=COLORS['azul_principal'], lw=2)
ax.set_title('17. Step Plot', fontstyle='italic')
ax.grid(True, alpha=0.3)

# 18. Barras Agrupadas
ax = fig2.add_subplot(gs2[1, 1])
grupos = ['Grupo A', 'Grupo B', 'Grupo C']
tel_vals = [0.45, 0.42, 0.35]
pres_vals = [0.40, 0.38, 0.30]
x_pos = np.arange(len(grupos))
width = 0.35
ax.bar(x_pos - width/2, tel_vals, width, label='Telef√≥nica', color=COLORS['azul_principal'])
ax.bar(x_pos + width/2, pres_vals, width, label='Presencial', color=COLORS['naranja'])
ax.set_xticks(x_pos)
ax.set_xticklabels(grupos)
ax.set_title('18. Barras Agrupadas', fontstyle='italic')
ax.legend()
ax.grid(True, alpha=0.3)

# 19. Hexbin Plot
ax = fig2.add_subplot(gs2[1, 2])
x_hex = np.random.normal(10, 3, 200)
y_hex = 0.6 - 0.02 * x_hex + np.random.normal(0, 0.1, 200)
hb = ax.hexbin(x_hex, y_hex, gridsize=10, cmap='Blues')
ax.set_title('19. Hexbin Plot', fontstyle='italic')

# 20. Contour Plot
ax = fig2.add_subplot(gs2[1, 3])
x_cont = np.linspace(-3, 3, 30)
y_cont = np.linspace(-3, 3, 30)
X, Y = np.meshgrid(x_cont, y_cont)
Z = np.exp(-(X**2 + Y**2)/2)
contours = ax.contour(X, Y, Z, colors=COLORS['azul_principal'])
ax.set_title('20. Contour Plot', fontstyle='italic')

# 21. Barras de Error
ax = fig2.add_subplot(gs2[2, 0])
x_err = ['A', 'B', 'C', 'D']
y_err = [20, 35, 30, 35]
yerr_vals = [2, 3, 4, 1]
ax.errorbar(x_err, y_err, yerr=yerr_vals, fmt='o', color=COLORS['azul_principal'], 
           capsize=5, capthick=2)
ax.set_title('21. Barras de Error', fontstyle='italic')
ax.grid(True, alpha=0.3)

# 22. Streamplot
ax = fig2.add_subplot(gs2[2, 1])
x_stream = np.linspace(-3, 3, 20)
y_stream = np.linspace(-3, 3, 20)
X_s, Y_s = np.meshgrid(x_stream, y_stream)
U = -1 - X_s**2 + Y_s
V = 1 + X_s - Y_s**2
ax.streamplot(X_s, Y_s, U, V, color=COLORS['azul_principal'], density=1.5)
ax.set_title('22. Streamplot', fontstyle='italic')

# 23. Quiver Plot
ax = fig2.add_subplot(gs2[2, 2])
x_q = np.arange(-2, 3)
y_q = np.arange(-2, 3)
X_q, Y_q = np.meshgrid(x_q, y_q)
U_q = X_q
V_q = Y_q
ax.quiver(X_q, Y_q, U_q, V_q, alpha=0.7, color=COLORS['azul_principal'])
ax.set_title('23. Quiver Plot', fontstyle='italic')

# 24. Barras 3D (simulado como barras con sombra)
ax = fig2.add_subplot(gs2[2, 3])
x_3d = ['X', 'Y', 'Z']
y_3d = [30, 25, 35]
bars = ax.bar(x_3d, y_3d, color=COLORS['azul_principal'])
# Simular efecto 3D con sombra
for i, bar in enumerate(bars):
    height = bar.get_height()
    ax.bar(i, height, bottom=-2, color=COLORS['gris_medio'], alpha=0.3, width=0.8)
ax.set_title('24. Efecto 3D', fontstyle='italic')
ax.grid(True, alpha=0.3)

# 25. Gr√°fico de Pareto
ax = fig2.add_subplot(gs2[3, 0])
causas = ['Causa A', 'Causa B', 'Causa C', 'Causa D']
frecuencias = [40, 30, 20, 10]
frecuencias_acum = np.cumsum(frecuencias)
total = sum(frecuencias)
porcentajes_acum = frecuencias_acum / total * 100

ax.bar(causas, frecuencias, color=COLORS['azul_principal'])
ax2 = ax.twinx()
ax2.plot(causas, porcentajes_acum, color=COLORS['naranja'], marker='o', linewidth=2)
ax2.set_ylabel('% Acumulado')
ax.set_title('25. Gr√°fico de Pareto', fontstyle='italic')
ax.grid(True, alpha=0.3)

# 26. Matriz de correlaci√≥n de variables principales
ax = fig2.add_subplot(gs2[3, 1])
data_corr = pd.DataFrame({
    'Edad': BASE_DATA['edades'],
    'Tiempo': BASE_DATA['tiempos'],
    'ROSC': BASE_DATA['outcomes'],
    'CPC': np.random.binomial(1, 0.3, 100)
})
corr = data_corr.corr()
im = ax.imshow(corr, cmap='Blues', vmin=-1, vmax=1)
ax.set_xticks(range(4))
ax.set_yticks(range(4))
ax.set_xticklabels(corr.columns)
ax.set_yticklabels(corr.columns)
ax.set_title('26. Matriz de correlaci√≥n', fontstyle='italic')

# 27. Gr√°fico de barras de odds ratios ajustados (multivariado)
ax = fig2.add_subplot(gs2[3, 2])
factores = ['Edad <65', 'Tiempo >mediana', 'RCP telef√≥nica']
or_vals = [1.4, 1.2, 1.8]
ci_l = [1.1, 0.9, 1.3]
ci_u = [1.8, 1.6, 2.5]
ax.barh(factores, or_vals, xerr=[np.array(or_vals)-np.array(ci_l), np.array(ci_u)-np.array(or_vals)], color=COLORS['azul_principal'], capsize=5)
ax.axvline(1, color=COLORS['gris_medio'], linestyle='--')
ax.set_xlabel('Odds Ratio (IC 95%)')
ax.set_title('27. OR ajustados multivariado', fontstyle='italic')

# 28. Gr√°fico de densidad de tiempo de llegada por grupo de RCP
ax = fig2.add_subplot(gs2[3, 3])
for i, (label, color) in enumerate(zip(['Telef√≥nica', 'Testigos', 'Sin RCP'], [COLORS['azul_principal'], COLORS['azul_claro'], COLORS['gris_medio']])):
    tiempos = BASE_DATA['tiempos'][i*33:(i+1)*33]
    sns.kdeplot(tiempos, ax=ax, label=label, color=color, fill=True, alpha=0.4)
ax.set_xlabel('Tiempo de llegada (min)')
ax.set_ylabel('Densidad')
ax.set_title('28. Densidad de tiempo de llegada', fontstyle='italic')
ax.legend()

plt.tight_layout()
plt.show()
```

```python
# Exportaci√≥n final: portada + galer√≠a en un solo PDF multip√°gina en la misma carpeta
import os
from matplotlib.backends.backend_pdf import PdfPages

# Carpeta destino: misma que este archivo
output_dir = os.path.dirname(os.path.abspath("1. design_language.ipynb"))
pdf_path = os.path.join(output_dir, "galeria_design_language_completa.pdf")

# Crear portada con jerarqu√≠a de color consistente
import matplotlib.pyplot as plt
fig_portada = plt.figure(figsize=(8.5, 11))
fig_portada.patch.set_facecolor(COLORS['blanco'])

# Eje invisible para dibujar l√≠neas
ax_portada = fig_portada.add_axes([0,0,1,1])
ax_portada.axis('off')

# T√≠tulo principal: azul_principal
fig_portada.text(0.5, 0.85, 'GALER√çA DE VISUALIZACIONES',
                 ha='center', fontsize=28, fontweight='bold', color=COLORS['azul_principal'], family='sans-serif')
# Subt√≠tulo: azul_oscuro
fig_portada.text(0.5, 0.78, 'Lenguaje de Dise√±o - RCP Transtelef√≥nica',
                 ha='center', fontsize=18, color=COLORS['azul_oscuro'], family='sans-serif')
# L√≠nea separadora: gris_medio
ax_portada.plot([0.2, 0.8], [0.76, 0.76], color=COLORS['gris_medio'], linewidth=3, transform=fig_portada.transFigure)
# Informaci√≥n del dise√±o: azul_medio
fig_portada.text(0.5, 0.70,
                 'Total: 25 tipos de gr√°ficos diferentes\n'
                 'Paleta: Azules/gris, naranja solo para destacar\n'
                 'Resoluci√≥n: 300 DPI para m√°xima calidad',
                 ha='center', fontsize=14, va='center', color=COLORS['azul_medio'], family='sans-serif')
# AVISO IMPORTANTE sobre mock data: naranja para destacar
fig_portada.text(0.5, 0.55, '‚ö†Ô∏è AVISO IMPORTANTE ‚ö†Ô∏è',
                 ha='center', fontsize=20, fontweight='bold',
                 color=COLORS['naranja'], bbox=dict(boxstyle="round,pad=0.5",
                 facecolor=COLORS['blanco'], edgecolor=COLORS['naranja'], linewidth=2))
fig_portada.text(0.5, 0.48,
                 'TODOS LOS DATOS SON SIMULADOS\n(MOCK DATA)',
                 ha='center', fontsize=16, fontweight='bold', color=COLORS['azul_oscuro'], family='sans-serif')
fig_portada.text(0.5, 0.40,
                 'Los datos utilizados en estas visualizaciones\n'
                 'son completamente simulados y NO representan\n'
                 'resultados reales del estudio RCP Transtelef√≥nica.',
                 ha='center', fontsize=12, va='center', color=COLORS['azul_oscuro'], family='sans-serif')
# Prop√≥sito del documento: azul_principal
fig_portada.text(0.5, 0.25,
                 'PROP√ìSITO DEL DOCUMENTO:',
                 ha='center', fontsize=14, fontweight='bold', color=COLORS['azul_principal'], family='sans-serif')
fig_portada.text(0.5, 0.20,
                 'Este documento sirve exclusivamente como referencia\n'
                 'de dise√±o visual para las futuras visualizaciones\n'
                 'del estudio con datos reales.',
                 ha='center', fontsize=12, style='italic', color=COLORS['gris_medio'], family='sans-serif')
# L√≠nea final: azul_principal
ax_portada.plot([0.2, 0.8], [0.12, 0.12], color=COLORS['azul_principal'], linewidth=2, transform=fig_portada.transFigure)

# Exportar PDF multip√°gina
with PdfPages(pdf_path) as pdf:
    pdf.savefig(fig_portada, bbox_inches='tight', dpi=300)
    pdf.savefig(fig1, bbox_inches='tight', dpi=300)
    pdf.savefig(fig2, bbox_inches='tight', dpi=300)
    plt.close(fig_portada)
    plt.close(fig1)
    plt.close(fig2)
print(f"‚úÖ PDF multip√°gina exportado: {pdf_path}")
```

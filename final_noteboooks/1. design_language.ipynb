Este notebook utiliza **datos simulados** (mock data) √∫nicamente con fines de visualizaci√≥n del dise√±o gr√°fico.

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# ====================================================================
# CONFIGURACI√ìN TIPOGR√ÅFICA ID√âNTICA AL LATEX
# ====================================================================

# Configuraci√≥n para coincidir exactamente con LaTeX \renewcommand{\familydefault}{\sfdefault}
plt.rcParams.update({
    # Tipograf√≠a sans serif como fuente principal (igual que LaTeX)
    'font.family': 'sans-serif',
    'font.sans-serif': ['DejaVu Sans', 'Arial', 'Liberation Sans', 'Helvetica'],
    'font.size': 10,              # Coincide con 10pt del LaTeX
    'axes.titlesize': 12,         # Proporcional
    'axes.labelsize': 10,         # Igual que tama√±o base
    'legend.fontsize': 9,         # Ligeramente menor
    'xtick.labelsize': 9,         # Menor para ejes
    'ytick.labelsize': 9,         # Menor para ejes
    'figure.titlesize': 14,       # Para t√≠tulos principales
    
    # Configuraci√≥n cient√≠fica compatible con LaTeX
    'mathtext.fontset': 'dejavusans',  # Sans serif para matem√°ticas
    'text.usetex': False,              # Sin LaTeX directo por compatibilidad
    
    # Calidad de renderizado profesional
    'figure.dpi': 300,
    'savefig.dpi': 300,
    'savefig.bbox': 'tight',
    'savefig.facecolor': 'white',
    'savefig.edgecolor': 'none',
    
    # Estilo limpio y profesional
    'axes.linewidth': 0.8,        # L√≠neas m√°s finas
    'grid.linewidth': 0.5,        # Grilla sutil
    'lines.linewidth': 1.5,       # L√≠neas de datos
    'patch.linewidth': 0.8,       # Bordes de patches
})

# ====================================================================
# PALETA CROM√ÅTICA SAMUR-PC CONSISTENTE
# ====================================================================

# Paleta principal SAMUR-PC con mayor consistencia
SAMUR_COLORS = {
    # Colores principales - tonos consistentes
    'rojo_principal': '#E53E3E',     # Rojo SAMUR principal
    'azul_principal': '#3182CE',     # Azul institucional
    'verde_principal': '#38A169',    # Verde cl√≠nico
    'violeta_principal': '#805AD5',  # Violeta anal√≠tico
    
    # Variaciones claras (para fondos)
    'rojo_claro': '#FED7D7',
    'azul_claro': '#BEE3F8', 
    'verde_claro': '#C6F6D5',
    'violeta_claro': '#E9D8FD',
    
    # Variaciones oscuras (para bordes/texto)
    'rojo_oscuro': '#C53030',
    'azul_oscuro': '#2A69AC',
    'verde_oscuro': '#2F855A',
    'violeta_oscuro': '#6B46C1',
    
    # Escala de grises neutra
    'negro': '#1A202C',
    'gris_oscuro': '#2D3748',
    'gris_medio': '#4A5568',
    'gris_claro': '#A0AEC0',
    'gris_muy_claro': '#F7FAFC',
}

# Paletas tem√°ticas simplificadas
PALETAS = {
    'principal': [
        SAMUR_COLORS['rojo_principal'],
        SAMUR_COLORS['azul_principal'], 
        SAMUR_COLORS['verde_principal'],
        SAMUR_COLORS['violeta_principal']
    ],
    'secuencial_rojo': [
        '#FED7D7', '#FCA5A5', '#F87171', 
        '#E53E3E', '#C53030', '#9B2C2C'
    ],
    'secuencial_azul': [
        '#BEE3F8', '#90CDF4', '#63B3ED',
        '#3182CE', '#2A69AC', '#1E4A8C'
    ],
    'clinical': [
        SAMUR_COLORS['verde_principal'],  # √âxito/supervivencia
        SAMUR_COLORS['violeta_principal'], # Neutro/seguimiento
        SAMUR_COLORS['rojo_principal']     # Cr√≠tico/mortalidad
    ]
}

print("üè• LENGUAJE DE DISE√ëO SAMUR-PC")
print("=" * 50)
print("üìä Tipograf√≠a: Sans serif (id√©ntica a LaTeX)")
print("üéØ Tama√±o base: 10pt (coincide con LaTeX)")
print("üé® Paleta: 4 colores principales + variaciones")
print("‚ú® Configuraci√≥n optimizada para consistencia")

# ====================================================================
# GRID EXPANDIDO DE GR√ÅFICOS √öTILES PARA EL ESTUDIO
# ====================================================================

fig = plt.figure(figsize=(20, 16))
gs = fig.add_gridspec(4, 4, hspace=0.6, wspace=0.5)

fig.suptitle('Lenguaje de dise√±o - Estudio RCP Transtelef√≥nica', 
             fontsize=16, fontweight='bold', color=SAMUR_COLORS['negro'], y=0.98)

# 1. Barras con intervalos de confianza
ax1 = fig.add_subplot(gs[0, 0])
categorias = ['RCP\nTelef√≥nica', 'RCP\nPresencial']
valores = [0.42, 0.38]
errores = [0.08, 0.07]

bars = ax1.bar(categorias, valores, yerr=errores, capsize=5,
               color=PALETAS['principal'][:2], alpha=0.8, 
               edgecolor=SAMUR_COLORS['negro'], linewidth=0.8)

for bar, valor in zip(bars, valores):
    height = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width()/2., height + 0.05,
             f'{valor:.2f}', ha='center', va='bottom', fontweight='bold')

ax1.set_ylabel('Tasa ROSC (IC 95%)', fontweight='bold')
ax1.set_title('1. Barras con IC', fontweight='bold')
ax1.set_ylim(0, 0.6)
ax1.grid(True, alpha=0.3)

# 2. L√≠neas temporales
ax2 = fig.add_subplot(gs[0, 1])
meses = np.arange(1, 13)
np.random.seed(42)
telefonica = np.random.normal(0.4, 0.05, 12)
presencial = np.random.normal(0.35, 0.04, 12)

ax2.plot(meses, telefonica, marker='o', linewidth=1.5, 
         color=SAMUR_COLORS['rojo_principal'], label='RCP Telef√≥nica')
ax2.plot(meses, presencial, marker='s', linewidth=1.5,
         color=SAMUR_COLORS['azul_principal'], label='RCP Presencial')

ax2.set_xlabel('Mes', fontweight='bold')
ax2.set_ylabel('Tasa ROSC', fontweight='bold')
ax2.set_title('2. Evoluci√≥n Temporal', fontweight='bold')
ax2.legend(fontsize=9)
ax2.grid(True, alpha=0.3)

# 3. Histograma con distribuci√≥n
ax3 = fig.add_subplot(gs[0, 2])
np.random.seed(123)
edades = np.random.normal(65, 15, 500)

ax3.hist(edades, bins=20, alpha=0.7, color=SAMUR_COLORS['verde_principal'],
         edgecolor=SAMUR_COLORS['negro'], linewidth=0.8)
ax3.axvline(np.mean(edades), color=SAMUR_COLORS['rojo_principal'], 
           linestyle='--', linewidth=1.5, label=f'Media: {np.mean(edades):.1f}')

ax3.set_xlabel('Edad (a√±os)', fontweight='bold')
ax3.set_ylabel('Frecuencia', fontweight='bold')
ax3.set_title('3. Distribuci√≥n Edad', fontweight='bold')
ax3.legend(fontsize=9)
ax3.grid(True, alpha=0.3)

# 4. Box plot comparativo
ax4 = fig.add_subplot(gs[0, 3])
data_box = [np.random.normal(8, 2, 100), np.random.normal(12, 3, 100)]
labels_box = ['Telef√≥nica', 'Presencial']

bp = ax4.boxplot(data_box, tick_labels=labels_box, patch_artist=True)
for patch, color in zip(bp['boxes'], PALETAS['principal'][:2]):
    patch.set_facecolor(color)
    patch.set_alpha(0.7)

ax4.set_ylabel('Tiempo (min)', fontweight='bold')
ax4.set_title('4. Box Plot Comparativo', fontweight='bold')
ax4.grid(True, alpha=0.3)

# 5. Gr√°fico de sectores (pie chart)
ax5 = fig.add_subplot(gs[1, 0])
outcomes = ['ROSC', 'No ROSC']
sizes = [35, 65]
colors_pie = [SAMUR_COLORS['verde_principal'], SAMUR_COLORS['gris_claro']]

wedges, texts, autotexts = ax5.pie(sizes, labels=outcomes, autopct='%1.1f%%',
                                  colors=colors_pie, startangle=90)
for autotext in autotexts:
    autotext.set_color('white')
    autotext.set_fontweight('bold')

ax5.set_title('5. Distribuci√≥n Outcomes', fontweight='bold')

# 6. Dispersi√≥n con regresi√≥n
ax6 = fig.add_subplot(gs[1, 1])
np.random.seed(456)
x_scatter = np.random.uniform(5, 20, 100)
y_scatter = 0.6 - 0.02 * x_scatter + np.random.normal(0, 0.1, 100)

ax6.scatter(x_scatter, y_scatter, alpha=0.6, 
           c=SAMUR_COLORS['violeta_principal'], s=50)

# L√≠nea de regresi√≥n
z = np.polyfit(x_scatter, y_scatter, 1)
p = np.poly1d(z)
ax6.plot(x_scatter, p(x_scatter), color=SAMUR_COLORS['rojo_principal'], 
         linewidth=1.5, linestyle='--')

ax6.set_xlabel('Tiempo Respuesta (min)', fontweight='bold')
ax6.set_ylabel('Prob. ROSC', fontweight='bold')
ax6.set_title('6. Correlaci√≥n Tiempo-Outcome', fontweight='bold')
ax6.grid(True, alpha=0.3)

# 7. Forest Plot (Odds Ratios)
ax7 = fig.add_subplot(gs[1, 2])
variables = ['Edad >65', 'Sexo M', 'Comorbilidad', 'Tiempo >10min']
odds_ratios = [1.2, 0.8, 1.5, 0.6]
ci_lower = [0.9, 0.6, 1.1, 0.4]
ci_upper = [1.6, 1.1, 2.0, 0.9]

y_pos = np.arange(len(variables))

# Plot points and error bars
for i, (or_val, lower, upper) in enumerate(zip(odds_ratios, ci_lower, ci_upper)):
    color = SAMUR_COLORS['verde_principal'] if or_val > 1 else SAMUR_COLORS['rojo_principal']
    ax7.errorbar(or_val, i, xerr=[[or_val-lower], [upper-or_val]], 
                fmt='o', color=color, capsize=5, capthick=1.5, markersize=6)

# Reference line at OR=1
ax7.axvline(x=1, color=SAMUR_COLORS['gris_oscuro'], linestyle='--', alpha=0.7)

ax7.set_yticks(y_pos)
ax7.set_yticklabels(variables)
ax7.set_xlabel('Odds Ratio (IC 95%)', fontweight='bold')
ax7.set_title('7. Forest Plot OR', fontweight='bold')
ax7.grid(True, alpha=0.3)

# 8. Heatmap de correlaciones
ax8 = fig.add_subplot(gs[1, 3])
np.random.seed(789)
variables_corr = ['Edad', 'Tiempo', 'ROSC', 'Supervivencia']
corr_matrix = np.random.rand(4, 4)
corr_matrix = (corr_matrix + corr_matrix.T) / 2  # Sim√©trica
np.fill_diagonal(corr_matrix, 1)

im = ax8.imshow(corr_matrix, cmap='RdYlBu_r', vmin=-1, vmax=1)
ax8.set_xticks(range(len(variables_corr)))
ax8.set_yticks(range(len(variables_corr)))
ax8.set_xticklabels(variables_corr, rotation=45)
ax8.set_yticklabels(variables_corr)

# A√±adir valores en el heatmap
for i in range(len(variables_corr)):
    for j in range(len(variables_corr)):
        text = ax8.text(j, i, f'{corr_matrix[i, j]:.2f}',
                       ha="center", va="center", color="black", fontweight='bold')

ax8.set_title('8. Matriz Correlaciones', fontweight='bold')

# 9. Violin Plot
ax9 = fig.add_subplot(gs[2, 0])
np.random.seed(101)
data_violin = [np.random.normal(65, 12, 200), 
               np.random.normal(70, 10, 180),
               np.random.normal(62, 15, 150)]
labels_violin = ['ROSC', 'No ROSC', '√âxitus']

parts = ax9.violinplot(data_violin, positions=[1, 2, 3], showmeans=True)
for pc, color in zip(parts['bodies'], PALETAS['clinical']):
    pc.set_facecolor(color)
    pc.set_alpha(0.7)

ax9.set_xticks([1, 2, 3])
ax9.set_xticklabels(labels_violin)
ax9.set_ylabel('Edad (a√±os)', fontweight='bold')
ax9.set_title('9. Violin Plot Edad-Outcome', fontweight='bold')
ax9.grid(True, alpha=0.3)

# 10. Barras agrupadas
ax10 = fig.add_subplot(gs[2, 1])
grupos_edad = ['<65', '65-75', '>75']
rosc_tel = [0.45, 0.42, 0.35]
rosc_pres = [0.40, 0.38, 0.30]

x = np.arange(len(grupos_edad))
width = 0.35

bars1 = ax10.bar(x - width/2, rosc_tel, width, label='Telef√≥nica',
                color=SAMUR_COLORS['rojo_principal'], alpha=0.8)
bars2 = ax10.bar(x + width/2, rosc_pres, width, label='Presencial',
                color=SAMUR_COLORS['azul_principal'], alpha=0.8)

ax10.set_xlabel('Grupo Edad', fontweight='bold')
ax10.set_ylabel('Tasa ROSC', fontweight='bold')
ax10.set_title('10. Barras Agrupadas', fontweight='bold')
ax10.set_xticks(x)
ax10.set_xticklabels(grupos_edad)
ax10.legend(fontsize=9)
ax10.grid(True, alpha=0.3)

# 11. √Årea bajo curva (AUC)
ax11 = fig.add_subplot(gs[2, 2])
fpr = np.linspace(0, 1, 100)
tpr = np.sqrt(fpr) + np.random.normal(0, 0.05, 100)
tpr = np.clip(tpr, 0, 1)

ax11.plot(fpr, tpr, color=SAMUR_COLORS['rojo_principal'], linewidth=1.5, 
         label='ROC Curve (AUC = 0.75)')
ax11.plot([0, 1], [0, 1], color=SAMUR_COLORS['gris_oscuro'], 
         linestyle='--', alpha=0.7, label='Random')
ax11.fill_between(fpr, tpr, alpha=0.2, color=SAMUR_COLORS['rojo_principal'])

ax11.set_xlabel('1 - Especificidad', fontweight='bold')
ax11.set_ylabel('Sensibilidad', fontweight='bold')
ax11.set_title('11. Curva ROC', fontweight='bold')
ax11.legend(fontsize=9)
ax11.grid(True, alpha=0.3)

# 12. Supervivencia Kaplan-Meier
ax12 = fig.add_subplot(gs[2, 3])
tiempo = np.linspace(0, 30, 100)
supervivencia_tel = np.exp(-tiempo/20) + np.random.normal(0, 0.02, 100)
supervivencia_pres = np.exp(-tiempo/18) + np.random.normal(0, 0.02, 100)
supervivencia_tel = np.clip(supervivencia_tel, 0, 1)
supervivencia_pres = np.clip(supervivencia_pres, 0, 1)

ax12.plot(tiempo, supervivencia_tel, color=SAMUR_COLORS['rojo_principal'], 
         linewidth=1.5, label='RCP Telef√≥nica')
ax12.plot(tiempo, supervivencia_pres, color=SAMUR_COLORS['azul_principal'], 
         linewidth=1.5, label='RCP Presencial')

ax12.set_xlabel('Tiempo (d√≠as)', fontweight='bold')
ax12.set_ylabel('Supervivencia', fontweight='bold')
ax12.set_title('12. Kaplan-Meier', fontweight='bold')
ax12.legend(fontsize=9)
ax12.grid(True, alpha=0.3)

# 13. Barras horizontales con ranking
ax13 = fig.add_subplot(gs[3, 0])
factores = ['Edad', 'Tiempo\nrespuesta', 'Comorbilidad', 'Ritmo\ninicial']
importancia = [0.8, 0.65, 0.45, 0.35]

bars_h = ax13.barh(factores, importancia, color=PALETAS['secuencial_rojo'][2:6], alpha=0.8)
for bar, valor in zip(bars_h, importancia):
    width = bar.get_width()
    ax13.text(width + 0.02, bar.get_y() + bar.get_height()/2,
             f'{valor:.2f}', ha='left', va='center', fontweight='bold')

ax13.set_xlabel('Importancia Relativa', fontweight='bold')
ax13.set_title('13. Ranking Factores', fontweight='bold')
ax13.grid(True, alpha=0.3)

# 14. Radial/Polar plot
ax14 = fig.add_subplot(gs[3, 1], projection='polar')
categorias_radar = ['ROSC', 'Supervivencia', 'CPC Favorable', 'Tiempo Respuesta']
valores_tel = [0.8, 0.6, 0.5, 0.7]
valores_pres = [0.7, 0.7, 0.6, 0.5]

angles = np.linspace(0, 2*np.pi, len(categorias_radar), endpoint=False).tolist()
valores_tel += valores_tel[:1]  # Cerrar el c√≠rculo
valores_pres += valores_pres[:1]
angles += angles[:1]

ax14.plot(angles, valores_tel, color=SAMUR_COLORS['rojo_principal'], 
         linewidth=1.5, label='Telef√≥nica', marker='o')
ax14.plot(angles, valores_pres, color=SAMUR_COLORS['azul_principal'], 
         linewidth=1.5, label='Presencial', marker='s')
ax14.fill(angles, valores_tel, alpha=0.25, color=SAMUR_COLORS['rojo_principal'])
ax14.fill(angles, valores_pres, alpha=0.25, color=SAMUR_COLORS['azul_principal'])

ax14.set_xticks(angles[:-1])
ax14.set_xticklabels(categorias_radar, fontsize=9)
ax14.set_ylim(0, 1)
ax14.set_title('14. Radar Chart\nComparativo', fontweight='bold', pad=20)
ax14.legend(fontsize=9, loc='upper right', bbox_to_anchor=(1.3, 1.0))

# 15. Stacked Bar (Barras apiladas)
ax15 = fig.add_subplot(gs[3, 2])
grupos_stack = ['Telef√≥nica', 'Presencial']
rosc_values = [35, 30]
no_rosc_values = [65, 70]

bars1 = ax15.bar(grupos_stack, rosc_values, color=SAMUR_COLORS['verde_principal'], 
                alpha=0.8, label='ROSC')
bars2 = ax15.bar(grupos_stack, no_rosc_values, bottom=rosc_values,
                color=SAMUR_COLORS['gris_claro'], alpha=0.8, label='No ROSC')

ax15.set_ylabel('Porcentaje', fontweight='bold')
ax15.set_title('15. Barras Apiladas', fontweight='bold')
ax15.legend(fontsize=9)
ax15.grid(True, alpha=0.3)

# 16. Dot plot
ax16 = fig.add_subplot(gs[3, 3])
hospitales = ['H1', 'H2', 'H3', 'H4', 'H5']
tasas_rosc = [0.42, 0.38, 0.45, 0.35, 0.40]
colores_dot = [SAMUR_COLORS['verde_principal'] if t > 0.4 else SAMUR_COLORS['rojo_principal'] 
               for t in tasas_rosc]

ax16.scatter(tasas_rosc, hospitales, c=colores_dot, s=100, alpha=0.8)
for i, (hospital, tasa) in enumerate(zip(hospitales, tasas_rosc)):
    ax16.text(tasa + 0.01, i, f'{tasa:.2f}', va='center', fontweight='bold')

ax16.axvline(x=0.4, color=SAMUR_COLORS['gris_oscuro'], linestyle='--', alpha=0.7)
ax16.set_xlabel('Tasa ROSC', fontweight='bold')
ax16.set_title('16. Dot Plot Hospitales', fontweight='bold')
ax16.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
